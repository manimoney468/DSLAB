```python
from sklearn.svm import SVC
import numpy as np

# Features
X = np.array([[3, 1], [3, -1], [6, 1], [6, -1], [1, 0], [0, 1], [0, -1], [-1, 0]])

# Labels
Y = np.array([1, 1, 1, 1, -1, -1, -1, -1])

# Fit SVM classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X, Y)

# Get the coefficients of the hyperplane
weights = svm_classifier.coef_[0]
bias = svm_classifier.intercept_[0]

# Print the coefficients of the hyperplane
print("Coefficients of the hyperplane:", weights)
print("Bias of the hyperplane:", bias)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Features
X = np.array([[3, 1], [3, -1], [6, 1], [6, -1], [1, 0], [0, 1], [0, -1], [-1, 0]])

# Labels
Y = np.array([1, 1, 1, 1, -1, -1, -1, -1])

# Fit SVM classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X, Y)

# Get the coefficients of the hyperplane
weights = svm_classifier.coef_[0]
bias = svm_classifier.intercept_[0]

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired, label='Data points')

# Plot the hyperplane
x_values = np.linspace(-1, 7, 10)
y_values = -(weights[0] * x_values + bias) / weights[1]
plt.plot(x_values, y_values, label='Discriminating hyperplane')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM: Discriminating Hyperplane')
plt.legend()
plt.grid(True)
plt.show()

```
